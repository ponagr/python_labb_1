```python
import matplotlib.pyplot as plt
# 1. Based on your solution code above, create a function that is able to take in new data files in similar format as dna_raw.txt with
#    different number of sequences, and produce the same results above
#Skicka in .txt filen
    #[x] ls den
    #[x] spara varje sekvens i en dictionary
    #[x] rkna varje bokstav i varje sekvens
    #[x] grafa varje sekvens

# 2. you have received also another DNA codes that is more complicated. The file is called
# dna_raw_complicated.txt. In this file, each DNA sequence can be composed of multiple lines of data in
# the text file. In this case, are you able to solve the task in the same way as before? If not, update your
# code to solve the same task with the new data

def calculate_dna(inserted_file):
    with open(inserted_file, "r") as file:
        opened_file = file.read()
    print(opened_file)
    splitted_file = [row.lower() for row in opened_file.split(">seq") if row]   #Skapa en lista som innehller det som finns i [xx] , appendar row om det finns en row
    # print(splitted_file)
    dna_list = []
    for row in splitted_file:
        parts = row.split()         #splitta p mellanslag (\n) som blir parts
        dna_dictionary = {
            "sequence_number": parts[0], 
            "dna_data": "".join(parts[1:]) 
        }
        dna_list.append(dna_dictionary)
    # print(dna_list)

    dna_list_counted = []
    i = 0
    for data in dna_list: 

        dna_data = data.get("dna_data")
        dna_dictionary_counted = {
            "sequence_number": data.get("sequence_number"),
            "A": dna_data.count("a"),
            "T": dna_data.count("t"),
            "C": dna_data.count("c"),
            "G": dna_data.count("g")
        }
        dna_list_counted.append(dna_dictionary_counted)
        graf_dna(dna_dictionary_counted, i, inserted_file)
        i += 1
    
    # print(dna_list_counted)
    
def graf_dna (dna: dict, i, file_name): 
    colors = ["yellow", "green", "blue", "red"]
    
    x = ["A", "T", "C", "G"]      #Bokstver
    y = [dna.get("A"), dna.get("T"), dna.get("C"),dna.get("G") ]      #vrden
    plt.title("sequence: " + dna.get("sequence_number"))
    plt.figtext(0.05, 0.9, file_name)
    plt.xlabel("X-Axel")
    plt.ylabel("Y-Axel")
    plt.bar(x,y, color=colors[i])
    plt.legend(["$Adenine$, $Thymine$, $Cytosine$, $Guanine$"])      #Ha denna efter bar, blir samma frg som staplarna r
    plt.show()


calculate_dna("dna_raw.txt")
print("---------------------------------------------------------------------")
calculate_dna("dna_raw_complicated.txt")
```
